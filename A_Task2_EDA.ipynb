```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tips = sns.load_dataset('tips')
```

```python
print("Tips dataset loaded successfully for EDA.")
print("\nFirst 5 rows of the DataFrame:")
print(df_tips.head())
```

```python
print("\nInitial DataFrame Info:")
df_tips.info()

print("\n")
```

```python
print("\n--- Data Structure & Types ---")
print("DataFrame Information (Columns, Non-Null Counts, Dtypes):")
df_tips.info()

print("\nDataFrame Shape (Rows, Columns):")
print(df_tips.shape)

print("\nColumn Names and their data types:")
print(df_tips.dtypes)
```

```python
print("\n--- Missing Values Check ---")
print("Number of missing values per column:")
print(df_tips.isnull().sum())

print("\nPercentage of missing values per column:")
print((df_tips.isnull().sum() / len(df_tips)) * 100)
```

```python
print("\n--- Descriptive Statistics for Numerical Columns ---")
print(df_tips.describe())

print("\n--- Value Counts for Categorical Columns ---")
for col in ['sex', 'smoker', 'day', 'time']:
    print(f"\nValue Counts for '{col}':")
    print(df_tips[col].value_counts())
```

```python
print("\n--- Initial Visualizations for Trends & Patterns ---")

# Set plot style
sns.set_style("whitegrid")

# 1. Distribution of Total Bill and Tip Amount
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df_tips['total_bill'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(df_tips['tip'], bins=20, kde=True, color='lightcoral')
plt.title('Distribution of Tip Amount')
plt.xlabel('Tip Amount ($)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 2. Relationship between Total Bill and Tip (Scatter Plot)
plt.figure(figsize=(10, 7))
sns.scatterplot(x='total_bill', y='tip', data=df_tips, hue='smoker', size='size', sizes=(20, 400), alpha=0.7)
plt.title('Total Bill vs. Tip Amount (Colored by Smoker, Sized by Party Size)')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip Amount ($)')
plt.grid(True)
plt.tight_layout()
plt.show()

# 3. Tip Amount by Day of the Week (Box Plot)
plt.figure(figsize=(8, 6))
sns.boxplot(x='day', y='tip', data=df_tips, palette='viridis', order=['Thur', 'Fri', 'Sat', 'Sun'])
plt.title('Tip Amount by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Tip Amount ($)')
plt.tight_layout()
plt.show()

# 4. Total Bill by Time of Day (Violin Plot)
plt.figure(figsize=(8, 6))
sns.violinplot(x='time', y='total_bill', data=df_tips, palette='pastel')
plt.title('Total Bill by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Total Bill ($)')
plt.tight_layout()
plt.show()

# 5. Tip Percentage Calculation and Distribution
df_tips['tip_percentage'] = (df_tips['tip'] / df_tips['total_bill']) * 100
plt.figure(figsize=(10, 6))
sns.histplot(df_tips['tip_percentage'], bins=30, kde=True, color='orange')
plt.title('Distribution of Tip Percentage')
plt.xlabel('Tip Percentage (%)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 6. Grouped Analysis: Average Tip Percentage by various categories
print("\nAverage Tip Percentage by Gender:")
print(df_tips.groupby('sex')['tip_percentage'].mean())

print("\nAverage Tip Percentage by Smoker Status:")
print(df_tips.groupby('smoker')['tip_percentage'].mean())

print("\nAverage Tip Percentage by Day:")
print(df_tips.groupby('day')['tip_percentage'].mean().sort_values(ascending=False))
```
